```python
import pandas as pd 
import numpy as np
import seaborn as sns; sns.set_theme()
import matplotlib.pyplot as plt
```

```python
all_data = pd.read_csv('all_data.csv')
all_data = all_data.drop(['Unnamed: 0'],axis=1)
all_data
```

```python
data = all_data[all_data.columns.difference(['Location', 'Period', 
                                             'Happiness Score'])]
data.insert(0,'Happiness Score',all_data['Happiness Score'])
data
```

```python
# correlation matrix
dcorr = data.corr()
# correlation heatmap
plt.figure(figsize=(11, 9),dpi=100)
sns.heatmap(data=dcorr, annot=True, fmt=".1f",
           mask=np.triu(np.ones_like(dcorr,dtype=np.bool)))
plt.savefig('heatmap.png')
```

From these plots, we can find these each variable has a strong linear relationship between response variable (Happiness Score). `30-70cancerChdEtc` and `mortalityRatePoisoning` have negative relationship between `Happiness Score`, and other variables' relationships between response variable are positive.
